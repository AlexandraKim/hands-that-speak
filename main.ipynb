```python
import numpy as np
import os, os.path, time
import cv2
import csv  

format='.jpg'
myDir = r"Dataset\ASL\test"

folders = os.listdir(myDir)

rows = []
for i in range(len(folders)): 
    sign = folders[i]
    images = os.listdir(myDir + "\\" + sign) 
    for j in range(len(images)): 
        imgName = images[j]
        imgc = cv2.imread(myDir + "\\" + sign + "\\" + imgName, 0) # reading in grayscale
        resized = cv2.resize(imgc, (200, 200), interpolation = cv2.INTER_AREA)
        
        row = np.array(resized.flatten(), dtype=object)
        row = np.insert(row, 0, str(sign), axis=0) # inserting label
        rows.append(row)
        
# writing to csv
with open('test_resized_200_200.csv', 'w', encoding='UTF8', newline='') as f:
    writer = csv.writer(f)

    # write the header
    writer.writerow(getScvHeader())

    # write the data
    writer.writerows(rows)          
```

```python
def getScvHeader():
    header = ['label']
    for x in range(40000): # because our images are of size 28 * 28
        header.append('pixel' + str(x + 1))
    return header
```
